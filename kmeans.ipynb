```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"AirTraffic.csv")
df.head()
```

```python
df1 = df.copy()
df1.duplicated().sum()
```

```python
df1 = df1.drop_duplicates(keep='first')
df1.duplicated().sum()
```

```python
df1 = df1.drop(columns = ['Activity Period','Operating Airline','Operating Airline IATA Code','Terminal','Boarding Area'])
df1.head()
```

```python
from feature_engine.outliers import Winsorizer
a = Winsorizer(capping_method = 'iqr',tail='both',fold = 0.05, variables = ['Passenger Count'])
df1 = a.fit_transform(df1)
plt.boxplot(df1['Passenger Count'])
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
b = OneHotEncoder(sparse_output = False)
encoded_df = b.fit_transform(df1[['GEO Region','Month']])
encdf = pd.DataFrame(encoded_df,columns = b.get_feature_names_out())
df1 = pd.concat([df1.drop(columns=['GEO Region','Month']),encdf],axis=1)
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
c = StandardScaler()
df1 = c.fit_transform(df1)
scaled_df = pd.DataFrame(df1,columns= c.get_feature_names_out())
scaled_df.head()
```

```python
scaled_df.isna().sum()
clean = scaled_df[~np.isnan(scaled_df).any(axis=1)]
```

```python
from sklearn.cluster import KMeans

TWSS = []
k = list(range(2,100))

for i in k:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(clean)
    TWSS.append(kmeans.inertia_)

TWSS
```

```python
plt.plot(k, TWSS, 'ro-')
plt.xlabel("No_of_Clusters") 
plt.ylabel("total_within_SS")
```

```python
from kneed import KneeLocator
kl = KneeLocator(range(2, 100), TWSS, curve='convex', direction = 'decreasing')
kl.elbow
plt.style.use("ggplot")
plt.plot(range(2, 100), TWSS)
plt.xticks(range(2, 100))
plt.ylabel("Interia")
plt.axvline(x = kl.elbow, color = 'r', label = 'axvline - full height', ls = '--')
plt.show() 
kl.elbow
```

```python
from sklearn import metrics
model = KMeans(n_clusters=16)
y_pred = model.fit_predict(clean)
mb = pd.Series(y_pred)
silhouette = metrics.silhouette_score(clean,mb)
silhouette
```
